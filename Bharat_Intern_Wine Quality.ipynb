# Outliers

```python
sns.boxplot(x=wine['residual sugar'])
```

```python
sns.boxplot(x=wine['total sulfur dioxide'])
```

```python
# Here I am using IQR to remove outliers. 
# IQR for price column
Q1 = wine['total sulfur dioxide'].quantile(0.25)
Q3 = wine['total sulfur dioxide'].quantile(0.75)
print('Q3',Q3)
print('Q1',Q1)
IQR = Q3 - Q1
print('IQR:',IQR)
upper_bound = Q3 + 1.5 * IQR
lower_bound = Q1 - 1.5 * IQR
print("upper_bound",upper_bound)
print("lower_bound",lower_bound)
```

```python
wine=wine[wine['total sulfur dioxide']<upper_bound]
wine=wine[wine['total sulfur dioxide']>lower_bound]
```

```python
sns.boxplot(x=wine['total sulfur dioxide'])
```

### input split

```python
# independent variable
X=wine.drop(['quality'],axis=1)
```

```python
# dependent variable
y=wine['quality']
```

```python
# see the dependent variable data distribution
y.value_counts()
```

see the target variable is imbalanced so we need to balance this column, for balance this column i use oversampling method

### oversampling

```python
from imblearn.over_sampling import SMOTE
```

```python
X.shape
```

```python
y.shape
```

```python
x,y=SMOTE().fit_resample(X,y)
```

```python
x.shape
```

```python
y.shape
```

```python
y.value_counts()
```

### Split the dataset 

```python
# splitting the dataset
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=101)
```

# Feature Scaling

```python
# performing transformation of data
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train=sc.fit_transform(X_train)
X_train
```

```python
X_test=sc.transform(X_test)
X_test
```

# Train the Model

```python
from sklearn.linear_model import LinearRegression
```

```python
lm=LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
print('Coefficients: ',lm.coef_)
```

```python
y_pred=lm.predict(X_train)
```

```python
# Importing MAE, rsq, MSE
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn import metrics
```

```python
# checking r2_score
r2 = r2_score(y_train,y_pred)
print(r2)
```

```python
predictions=lm.predict(X_test)
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```
